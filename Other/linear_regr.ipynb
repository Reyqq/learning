# Линейная регрессия

## Введение в линейную регрессию

### Определение линейной регрессии

**Линейная регрессия** — это метод статистического анализа, используемый для определения зависимости между одной зависимой переменной (целевой) и одной или несколькими независимыми переменными (предикторами). Основная цель линейной регрессии заключается в нахождении уравнения, которое описывает, как целевая переменная изменяется в зависимости от значений предикторов. При этом предполагается, что связь между переменными носит линейный характер, то есть зависимость можно описать прямой линией на плоскости или гиперплоскостью в многомерном пространстве.

### Цель линейной регрессии

**Главная цель линейной регрессии** — предсказать значение целевой переменной на основе известных значений предикторов. Это достигается путем подбора таких коэффициентов в уравнении линейной регрессии, которые минимизируют разницу между предсказанными и фактическими значениями целевой переменной. Важно также выявить и проанализировать взаимосвязи между переменными, что может помочь в прогнозировании будущих данных или в интерпретации влияния предикторов.

### Примеры применения в реальном мире

Линейная регрессия широко используется в самых разных областях:

  * **Экономика:** прогнозирование доходов или расходов в зависимости от различных факторов (например, уровня образования, стажа работы и т.д.).

  * **Медицина:** оценка влияния различных факторов на здоровье человека (например, связь между курением и продолжительностью жизни).

  * **Маркетинг:** анализ продаж и выявление факторов, влияющих на рост спроса.
  
  * **Наука:** установление связи между изменениями температуры и уровня углекислого газа в атмосфере.

## Математическая модель линейной регрессии



### Уравнение линейной регрессии

Математическая модель линейной регрессии описывается следующим уравнением:

$y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \dots + \beta_n x_n + \epsilon$


где:
- \\(y\\) — целевая переменная,
- \\(x_1, x_2, \dots, x_n\\) — независимые переменные (предикторы),
- \\(\beta_0\\) — свободный член,
- \\(\beta_1, \beta_2, \dots, \beta_n\\) — коэффициенты,
- \\(\epsilon\\) — ошибка модели.


### Объяснение параметров модели (коэффициенты и свободный член)

- **Свободный член \\(\beta_0\\)** — это значение целевой переменной
\\(y\\), когда все предикторы равны нулю. Он задает точку пересечения модели с осью \\(y\\).

- **Коэффициенты \\(\beta_1, \beta_2, \dots, \beta_n\\)** — это параметры, которые показывают, на сколько изменится \\(y\\) при изменении соответствующего
\\(x_i\\) на единицу, при прочих равных. Эти коэффициенты указывают на силу и направление связи между предикторами и целевой переменной.


### Понятие ошибки и функции потерь

Модель линейной регрессии никогда не будет идеально предсказывать значения, так как существуют случайные факторы или ошибки измерения. Эти отклонения называются ошибками модели \\(\epsilon\\). Чтобы минимизировать эти ошибки, используется функция потерь, чаще всего — сумма квадратов ошибок:

$$
L(\beta) = \sum_{i=1}^{m} (y_i - \hat{y}_i)^2
$$

где:
- \\(y_i\\) — фактическое значение,
- \\(\hat{y}_i\\) — предсказанное значение,
- \\(m\\) — количество наблюдений.

Функция потерь показывает, насколько предсказанные значения отклоняются от фактических данных.

## Метод наименьших квадратов (МНК)

### Принцип работы МНК

**Метод наименьших квадратов (МНК)** — это метод оптимизации, применяемый для нахождения таких параметров \\(\beta\\), которые минимизируют сумму квадратов ошибок между фактическими значениями и предсказанными моделью. Основная цель МНК — подобрать такие коэффициенты, чтобы кривая, построенная на основании предсказанных значений, была как можно ближе к реальным данным.

**Формулировка задачи оптимизации**

Оптимизация задачи линейной регрессии через МНК заключается в минимизации следующей функции:

$$
L(\beta) = \| X \beta - y \|^2
$$

где:
- \\(X\\) — матрица признаков,
- \\(y\\) — вектор целевых значений,
- \\(\hat{\beta}\\) — вектор оценок коэффициентов.
